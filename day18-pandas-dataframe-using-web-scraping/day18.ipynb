## There are 197 page so we haveto run it 197 times with the help of loop

```python
final = pd.DataFrame()
for j in range(1, 2):
    # url='https://9animetv.to/filter?keyword=&type=&status=all&genre=&season=&language=&year=&sort=all&page={}'.format(j)
    webpage_9animetv = requests.get('https://9animetv.to/filter?keyword=&type=&status=all&genre=&season=&language=&year=&sort=all&page={}'+str(j)).text
    soup = BeautifulSoup(webpage_9animetv,'lxml')
    anime=soup.find_all('div',class_="flw-item item-qtip")
    name=[]
    ep=[]
    lan=[]

    for i in anime:
        name.append(i.find('h3').text.strip())
        ep.append(i.find('div',class_="tick-eps").text.strip())
        lan.append(i.find('div',class_="tick ltr").text.strip())


d={'name':name,'ep':ep,'lan':lan}
df=pd.DataFrame(d)

    
final=final._append(df,ignore_index=True)
```

```python
final = pd.DataFrame()
for j in range(1, 197):
    webpage_9animetv = requests.get('https://9animetv.to/filter?keyword=&type=&status=all&genre=&season=&language=&year=&sort=all&page='+str(j)).text
    soup = BeautifulSoup(webpage_9animetv,'lxml')
    anime=soup.find_all('div',class_="flw-item item-qtip")
    name=[]
    ep=[]
    lan=[]
    for i in anime:
        name.append(i.find('h3').text.strip())
        ep.append(i.find('div',class_="tick-eps").text.strip())
        lan.append(i.find('div',class_="tick ltr").text.strip())

    d={'name':name,'ep':ep,'lan':lan}
    df=pd.DataFrame(d)
    final=final._append(df,ignore_index=True)
    
```
